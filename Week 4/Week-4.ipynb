## Problem 1

Create a line plot of ZN and INDUS in the housing data.

1. For ZN, use a solid green line. For INDUS, use a blue dashed line.

2. Change the figure size to a width of 12 and height of 8.

3. Change the style sheet to something you find https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html

```python
from IPython.display import HTML

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.rcParams['figure.figsize'] = (12, 8)
```

```python
housing_csv = 'data/boston_housing_data.csv'
housing = pd.read_csv(housing_csv)
```

```python
plt.style.use('bmh')
housing[['ZN', 'INDUS']].plot(figsize=(12,8), style={'ZN': 'g', 'INDUS': '--b'});
```

## Problem 4

Create a scatter plot of two heatmap entries that appear to have a very positive correlation.

```python
feature1 = np.random.randn(1, 500)
feature2 = np.random.randn(1, 500)
feature3 = feature1 + feature2;
```

```python
data = {
    'feature1': feature1[0],
    'feature2': feature2[0],
    'feature3': feature3[0]
}
```

```python
data_frame = pd.DataFrame(data)
plt.style.use('fivethirtyeight')
data_frame.plot(kind='scatter', x='feature1', y='feature2', c='feature3', colormap='Blues')
```

## Problem 5

Create a scatter plot of two heatmap entries that appear to have negative correlation.

```python
feature4 = np.random.randn(1, 500)
feature5 = np.random.randn(1, 500)
feature6 = -1 * (feature4 + feature5);
```
